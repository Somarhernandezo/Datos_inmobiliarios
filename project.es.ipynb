```python
poblaciones = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
colores = ['blue', 'orange', 'green', 'purple']

plt.figure(figsize=(10,6))

for poblacion, color in zip(poblaciones, colores):
    
    precios = df_cinturon_sur[df_cinturon_sur['level5'] == poblacion]['price']
    
    
    precios_norm = (precios - precios.min()) / (precios.max() - precios.min())
    
    
    plt.hist(precios_norm, bins=10, alpha=0.5, color=color, label=poblacion, edgecolor='black')

plt.title("Histogramas de precios normalizados por población - Cinturón Sur")
plt.xlabel("Precio normalizado")
plt.ylabel("Cantidad de casas")
plt.legend()
plt.grid(axis='y', alpha=0.7)
plt.show()
```

```python

from ipyleaflet import Map, CircleMarker, basemaps
from ipywidgets import HTML

ds = pd.read_csv('assets/real_estate.csv', sep=';')

poblaciones_sur = ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]
df_cinturon_sur = ds[ds['level5'].isin(poblaciones_sur)]

df_cinturon_sur = df_cinturon_sur.dropna(subset=['latitude', 'longitude'])

coordenadas_dict = {}
for i, row in df_cinturon_sur.iterrows():
    coordenadas_dict[i] = {
        'poblacion': row['level5'],
        'lat': row['latitude'],
        'lon': row['longitude'],
        'address': row['address'],
        'price': row['price'],
        'rooms': row['rooms'],
        'bathrooms': row['bathrooms'],
        'surface': row['surface']
    }

colores = {
    "Fuenlabrada": "blue",
    "Leganés": "red",
    "Getafe": "green",
    "Alcorcón": "purple"
}


mapa = Map(center=(40.35, -3.8), zoom=11.5, basemap=basemaps.OpenStreetMap.Mapnik)

for info in coordenadas_dict.values():
    color = colores.get(info['poblacion'], 'gray')
    marker = CircleMarker(
        location=(info['lat'], info['lon']),
        radius=5,
        color=color,
        fill_color=color,
        fill_opacity=0.6
    )
  
    marker.popup = HTML(
        f"<b>{info['poblacion']}</b><br>"
        f"{info['address']}<br>"
        f"Precio: {info['price']} USD<br>"
        f"Habitaciones: {info['rooms']}, Baños: {info['bathrooms']}, Superficie: {info['surface']} m²"
    )
    mapa.add_layer(marker)


mapa
```
